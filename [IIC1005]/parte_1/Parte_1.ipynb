### Pregunta 3: Gráficos de torta
En esta parte queremos observar las distribuciones de las personas que vieron, están viendo o están por ver (Completed, Watching, Plan to Watch) los mejores 5 animes. 


Para poder encontrar los mejores 5 animes, desde pandas usamos la función sort() para ordenarlos por "Score" y la función head() elegir los 5 primeros desde el mayor al menor (ascending=False para orden descendiente).

Una vez creado el data frame, creamos los gráficos de torta (kind.'pie') usando el comando groupby(), para así agruparlos según el nombre y según el tipo (eje y). Además se usaron otras funciones como poner el título (title), que salgan los porcentajes (%1.1f%%) y asignar el tamaño del gráfico (figsize).

Luego se hará lo mismo pero los datos se harán agregados, para esto agruparemos por el nombre mediante groupby y las personas que vieron, están viendo o están por ver (Completed, Watching, Plan to Watch) según cada anime se hará el gráfico 

Información obtenida de: https://kontext.tech/column/code-snippets/402/pandas-dataframe-plot-pie-chart 

````python
anime_top = anime.sort_values(by="Score", ascending=False).head(5) #Obtener los 5 mejores animes (en base a puntaje)
````

````python
anime_top
````

````python
anime_top.groupby(['Name']).sum().plot(kind='pie', y='Watching',autopct="%1.1f%%", title="Watching Comparison", figsize=(7,7), legend=False)
````

````python
anime_top.groupby(['Name']).sum().plot(kind='pie', y='Completed',autopct="%1.1f%%", title="Completed Comparison", figsize=(7,7), legend=False)
````

````python
anime_top.groupby(['Name']).sum().plot(kind='pie', y='Plan to Watch',autopct="%1.1f%%", title="Plan to Watch Comparison", figsize=(7,7), legend=False)
````

### Pregunta 4: Regresión
En esta sección queremos realizar una regresión para conocer el comportamiento de los scores e interacciones desde el año 2000. Se mostrarán los gráficos mediante "Score", "Completed", "Plan to Watch" y "Watching". Para esto 'cortaremos' el data frame seleccionando los que son del año 2000 en adelante. Para hacer esto, transformaremos los datos de la columna "Year" a int para así poder ordenarlos y poder realizar las regresiones con el comando 'regplot'.

Para poder transformarlos a int se usó la siguiente información: https://datatofish.com/string-to-integer-dataframe/

Debido a que en la columna "Year" habían años escritos como 'Unknown', no se puede transformar a int. Para lograr esto, se modificará el DataFrame, haciendo que todos los 'Unknown' sean 0. Esto permitirá ordenar y seleccionar por años.

````python
anime.loc[anime["Year"]=="Unknown", "Year"]=0
anime['Year'] = anime['Year'].astype(int)
anime
````

````python
anime_2000=anime.loc[anime["Year"]>1999]
anime_2000
````

````python
anime_2000_final=anime_2000.groupby('Year').mean().drop(labels=['MAL_ID', 'Episodes'], axis=1)
anime_2000_final
````

````python
#https://datatofish.com/string-to-integer-dataframe/
````

````python
fig, ax = plt.subplots()
sns.regplot(data=anime_2000.sample(100),x='Year', y='Score', ax=ax, order=3, ci=None).set_title("Score means by year")
plt.xticks(rotation=45)
plt.show()
````

````python
fig, ax = plt.subplots()
sns.regplot(data=anime_2000.sample(100),x='Year', y='Watching', ax=ax, order=3, ci=None).set_title("Watching means by year")
plt.xticks(rotation=45)
plt.show()
````

````python
fig, ax = plt.subplots()
sns.regplot(data=anime_2000.sample(100),x='Year', y='Completed', ax=ax, order=3, ci=None).set_title("Completed means by year")
plt.xticks(rotation=45)
plt.show()
````

````python
fig, ax = plt.subplots()
sns.regplot(data=anime_2000.sample(100),x='Year', y='Plan to Watch', ax=ax, order=3, ci=None).set_title("Plan to Watch means by year")
plt.xticks(rotation=45)
plt.show()
````
